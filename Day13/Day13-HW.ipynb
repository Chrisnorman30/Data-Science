```python
import matplotlib.pyplot as plt

# Plot both lines on the same y-axis
ax = DF.plot(x='date', y='count_mask', label='Mask Counties', marker='o')
DF.plot(x='date', y='count_nomask', label='No Mask Counties', marker='o', ax=ax)

# Add titles and labels
plt.title('Kansas COVID-19 7-Day Rolling Average of Daily Cases Per 100K Population')
plt.xlabel('Date')
plt.ylabel('Cases per 100K Population')
plt.grid(True)
plt.show()
```

In this graph, it’s easy to see the real difference between mask and no-mask counties because both lines use the same y-axis. It shows that mask counties started higher but went down over time, while no-mask counties stayed lower but didn’t drop much. From the data and this visualization, we can see that in the time period shown, counties with mask mandates started with higher COVID case numbers but generally declined over time, while counties without mask mandates stayed lower but didn’t decrease much. This suggests that cases can change differently in different counties, but the graph alone **does not prove that masks caused the decline**—it only shows trends. The visualization aligns with the general understanding that mask-wearing may help reduce spread, but it does **not establish cause and effect**, so we should be careful about drawing strong conclusions from it.


Yes, there could definitely be confounding factors. Things like population density, testing rates, other public health measures, mobility, or even differences in reporting could affect COVID case trends in each county.
